```python
import cv2 
import mediapipe as mp 
```

```python
mp_face_dect=mp.solutions.face_detection
mp_draw=mp.solutions.drawing_utils
```

```python
mp_drawing=mp.solutions.drawing_utils
mp_drawing_style=mp.solutions.drawing_styles
mp_hand=mp.solutions.hands
```

```python
cap=cv2.VideoCapture(0)
p = mp_hand.Hands(model_complexity=0,min_tracking_confidence=0.5,min_detection_confidence=0.5) 
face_dect=mp_face_dect.FaceDetection(min_detection_confidence=1,model_selection=0) 
while cap.isOpened(): 
    r,frame=cap.read()
    frame=cv2.flip(frame,1) 
    
    frame=cv2.cvtColor(frame,cv2.COLOR_BGR2RGB)

    result=face_dect.process(frame)

    result1= p.process(frame)

    frame=cv2.cvtColor(frame,cv2.COLOR_RGB2BGR)

    if r== True: 
        for i in result.detections: 
            mp_draw.draw_detection(frame,i) 


        if result1.multi_hand_landmarks: 
            for landmarks in result1.multi_hand_landmarks: 
                mp_drawing.draw_landmarks(frame,landmarks,mp_hand.HAND_CONNECTIONS) 

        cv2.imshow("sid",frame)
        if cv2.waitKey(25) & 0xff==ord("q"):
            break
            
    else: 
        break
                
cap.release()
cv2.destroyAllWindows()
```
