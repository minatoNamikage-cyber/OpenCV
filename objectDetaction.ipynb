```python
import cv2
import numpy as np

def nothing(x):
    pass

cv2.namedWindow("siddharth")
cv2.createTrackbar("th", "siddharth", 0, 255, nothing)
cv2.createTrackbar("lb", "siddharth", 0, 255, nothing)
cv2.createTrackbar("lg", "siddharth", 0, 255, nothing)
cv2.createTrackbar("lr", "siddharth", 0, 255, nothing)
cv2.createTrackbar("hb", "siddharth", 255, 255, nothing)
cv2.createTrackbar("hg", "siddharth", 255, 255, nothing)
cv2.createTrackbar("hr", "siddharth", 255, 255, nothing)

cap = cv2.VideoCapture(0)

while cap.isOpened():
    r, frame = cap.read()
    if r == True:
        thr = cv2.getTrackbarPos("th", "siddharth")
        LB = cv2.getTrackbarPos("lb", "siddharth")
        LG = cv2.getTrackbarPos("lg", "siddharth")
        LR = cv2.getTrackbarPos("lr", "siddharth")
        HB = cv2.getTrackbarPos("hb", "siddharth")
        HG = cv2.getTrackbarPos("hg", "siddharth")
        HR = cv2.getTrackbarPos("hr", "siddharth")

        lower = np.array([LB, LG, LR])
        upper = np.array([HB, HG, HR])

        frame = cv2.flip(frame, 1)
        frame = cv2.resize(frame, (400, 400))
        hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)

        m = cv2.inRange(hsv, lower, upper)
        res = cv2.bitwise_and(frame, frame, mask=m)
        fr = cv2.bitwise_not(res)

        _, thi = cv2.threshold(m, thr, 255, cv2.THRESH_BINARY)
        contours_info = cv2.findContours(thi.copy(), cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
        cnt = contours_info[0] if len(contours_info) == 2 else contours_info[1]
        
        cv2.drawContours(frame, cnt, -1, (255, 0, 0), 2)

        cv2.imshow("thr", thi)
        cv2.imshow("res", res)
        cv2.imshow("mask", m)
        cv2.imshow("sid", frame)

        if cv2.waitKey(25) & 0xff == ord("q"):
            break
    else:
        break

cap.release()
cv2.destroyAllWindows()
```
